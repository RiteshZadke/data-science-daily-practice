Q1. Take a list of numbers and find the sum without using sum().

```python
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
total = 0
for i in numbers:
  total += i

print("Sum of List:",total)
```

Q2. Find the largest and smallest element in a list (without using max() or min()).

```python
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
small = numbers[0]
large = numbers[0]
for i in numbers:
  if i > large:
    large = i
  if i<small:
    small = i

print(f'Largest number: {large}\nSmallest number: {small}')
```

Q3. Count how many even and odd numbers are in a list.

```python
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
even = 0
odd = 0
for i in numbers:
  if i%2 == 0:
    even += 1
  else:
    odd += 1

print(f'Odd number count: {odd}\nEven numbrs count: {even}')
```

Q4. Remove duplicate elements from a list (preserve original order).

```python
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
unique = []
for num in numbers:
  if num not in unique:
    unique.append(num)
print(f"List without duplicates: {unique}")
```

Q5. Reverse a list without using reverse() or slicing.

```python
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
left = 0
right = len(numbers) - 1

while left < right:
  numbers[left], numbers[right] = numbers[right], numbers[left]
  left += 1
  right -= 1

print(f'Reverse list: {numbers}')
```

Q6. Find the second largest number in a list.

```python
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
large = 0
second_large = 0
for num in numbers:
  if num > large:
    second_large = large
    large = num

print(f"Second largest number: {second_large}")
```

Q10. Find the most frequent element in a list.

```python
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
freq = {}
for num in numbers:
    freq[num] = freq.get(num, 0) + 1

most_freq = numbers[0]
for num in freq:
    if freq[num] > freq[most_freq]:
        most_freq = num


print(f"Most frequent {most_freq}")
```
