Q1. Print numbers from 1 to 10 using a for loop.

```python
for i in range(1,11):
  print(i)
```

Q2. Print even numbers between 1 and 20.

```python
for i in range(1,21):
  if i % 2 == 0:
    print(i)
```

 Q3. Print the sum of numbers from 1 to n, where n is taken as input.

```python
n = int(input("Enter number: "))
total = 0
for i in range(1,n+1):
  total+=i
print(f'Sum of numbers form 1 to {n}: {total}')
```

Q4. Print the multiplication table of a given number (up to 10).

```python
num = int(input("Enter number: "))
for i in range(1,11):
  print(f'{i} x {num} = {i*num}')
```

Q5. Count the number of digits in a given integer.

```python
num = int(input("Enter number: "))
count = 0
for i in str(num):
  count +=1
print('Number of digit: ',count)
```

Q6. Reverse a number using a loop.

```python
n = int(input("Enter a number: "))
reverse = 0

while n > 0:
    digit = n % 10
    reverse = reverse * 10 + digit
    n = n // 10

print("Reversed number:", reverse)
```

Q7. Check whether a number is prime or not.

```python
num = int(input("Enter number: "))

if num <= 1:
    print("Not Prime")
else:
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime")
    else:
        print("Not Prime")
```

Q8. Print the factorial of a given number.

```python
num = int(input("Enter number: "))
if num < 0:
    print("Factorial is not defined for negative numbers")
else:
    fact = 1
    while num > 1:
        fact *= num
        num -= 1
    print(f'factorial is {fact}')
```

Q9. Print the following pattern:

```python
num = int(input("Enter number of layer: "))
for i in range(1,num+1):
  print('*'*i)
```

Q10. Given a number n, print the Fibonacci series up to n terms.

```python
n = int(input("Enter number: "))

a, b = 0, 1

for i in range(n):
    print(a)
    a, b = b, a + b
```
