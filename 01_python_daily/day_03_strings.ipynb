Q1. Take a string as input and print its length (without using len()).

```python
string = input("Enter any String: ")
length = 0
for i in string:
  length +=1

print('Length of string is',length)
```

Q2. Count the number of vowels in a given string.

```python
string = input('Enter any string: ')
counter = 0
for ch in string:
  if ch in 'aeiouAEIOU':
    counter += 1
if counter == 0:
  print("There is no vowels in string")
else:
  print("Vowels count is",counter)
```

Q3. Reverse a string without using slicing.

```python
string = input("Enter any string: ")
rev =''
for ch in string:
  rev = ch + rev
print("Reversed string: ",rev)
```

Q4. Check whether a string is a palindrome.

```python
string = input("Enter any string: ").lower()
rev = ""

for ch in string:
    rev = ch + rev

if rev == string:
    print("Palindrome")
else:
    print("Not Palindrome")
```

Q5. Count the frequency of each character in a string.

```python
string = input("Enter any string: ")
dic = {}
for ch in string.lower():
  if ch not in dic:
    dic[ch] = 1

  else :
    dic[ch] += 1
print('Character frequency:',dic)
```

Q6. Convert all uppercase letters to lowercase and vice versa
   (without using .swapcase()).

```python
string = input('Enter string:')
swap_case =''
for ch in string:
  if ch == ch.upper():
    swap_case += ch.lower()
  elif ch == ch.lower():
    swap_case += ch.upper()
  else:
    swap_case += ch
print("Swapcase string:",swap_case)
```

Q7. Remove all spaces from a string.

```python
string = input("Enter string: ")
result = ""

for ch in string:
    if ch != " ":
        result += ch

print("String without spaces:", result)
```

Q8. Find the first non-repeating character in a string.

```python
string = input('Enter String: ').lower()
dic = {}
for ch in string:
  if ch not in dic:
    dic[ch] = 1
  else:
    dic[ch] += 1

for ch in string:
  if dic[ch] == 1:
    print("first non-repeating character:",ch)
    break
else:
  print('No non-repeating character found')
```

Q9. Check whether two strings are anagrams of each other.

```python
string1 = input("Enter first string :").replace(" ", "").lower()
string2 = input("Enter second string :").replace(" ", "").lower()

if sorted(string1) == sorted(string2):
  print('Strigs are anagrams')

else:
  print('Strigs are not anagrams')
```

Q10. Find the longest word in a sentence.

```python
sen = input('Enter sentence: ')
sen = sen.split()
longest = ''
for word in sen:
  if len(word)>len(longest):
    longest = word
print('Longest word:',longest)
```
