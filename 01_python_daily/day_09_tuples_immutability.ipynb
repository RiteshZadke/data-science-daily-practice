Q5. Single Element Tuple Trap

Create:

* a tuple with one element

* a normal variable with parentheses

Print their types and explain the difference in comments.

```python
# tuple with one element
t = (10,)

# normal variable with parentheses
x = (10)

print(type(t))  # <class 'tuple'>
print(type(x))  # <class 'int'>
```

```python
# Explanation:
# (10,) is a tuple because of the trailing comma
# (10) is just an integer in parentheses, not a tuple
# The comma, not the parentheses, defines a tuple in Python
```

Q6. Tuple Inside List (Mutability Twist)

Create a list that contains a tuple.

Example:

data = [(1, 2), (3, 4)]


Try:

* modifying the list

* modifying the tuple

Explain what is allowed and why.

```python
data = [(1, 2), (3, 4)]

data.append((5, 6))
data[0] = (10, 20)

# modifying the tuple (not allowed)
data[1][0] = 30  # TypeError
```

```python
# Explanation:
# The list is mutable, so we can add, remove, or replace its elements.
# The tuple inside the list is immutable, so its elements cannot be changed.
# We can replace the entire tuple in the list, but we cannot modify values inside the tuple.
```

Q7. Given a tuple:

nums = (1, 2, 3, 2, 4, 2)


Find:

* how many times 2 appears

* index of first occurrence of 2

```python
nums = (1, 2, 3, 2, 4, 2)

# count how many times 2 appears
count_2 = nums.count(2)

# index of first occurrence of 2
first_index_2 = nums.index(2)

print(count_2)
print(first_index_2)
```

```python
# Explanation:
# count() returns the total number of occurrences of a value in the tuple
# index() returns the index of the first occurrence of the value
```

Q9. Given a tuple of numbers, calculate:

```python
nums = (10, 20, 30, 40)

# sum of elements
total = sum(nums)

# count of elements
count = len(nums)

print(total)
print(count)
```

Q10. Create a dictionary where:

* keys are tuples (x, y)

* values are distances from origin

```python
# dictionary with tuple keys and distance from origin as values
points = {
    (1, 2): (1**2 + 2**2) ** 0.5,
    (3, 4): (3**2 + 4**2) ** 0.5,
    (0, 5): (0**2 + 5**2) ** 0.5
}
print(points)
```

```python
# Explanation:
# Tuple (x, y) is used as a key because tuples are immutable and hashable
# Distance from origin is calculated using: √(x² + y²)
# Lists cannot be keys, but tuples can be
```
