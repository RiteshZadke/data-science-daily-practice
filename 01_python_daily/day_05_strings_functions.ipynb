Q1. Take a string input and count the number of vowels in it.

```python
def vowels(string):
  vow = 0
  for i in string:
    if i in 'aeiouAEIOU':
      vow += 1
  return vow
```

```python
vowels('ritesh')
```

Q2. Reverse a string without using slicing ([::-1]).

```python
def reverse(string):
  rev = ''
  for i in string:
    rev = i+rev

  return rev
```

```python
reverse("python")
```

Q3. Check whether a string is a palindrome.

```python
def is_palindrome(string):
  rev = ''
  count = False
  for i in string.lower():
    rev = i+rev

  if rev == string.lower():
   count = True
  else:
    count = False

  return count
```

```python
is_palindrome('mom')
```

Q4. Count the frequency of each character in a string.

```python
def char_freq(string):
  freq = {}
  for i in string:
    if i not in freq:
      freq[i] = 1
    else:
      freq[i] += 1

  return freq
```

```python
char_freq('frequency')
```

Q5. Find the first non-repeating character in a string.

```python
def non_rep(s):
    freq = {}

    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1

    for ch in s:
        if freq[ch] == 1:
            return ch

    return None
```

```python
non_rep('Python')
```

Q6. Write a function that takes a string and returns it in title case
   (example: "data science" → "Data Science").

```python
def title_case(string):
    result = []
    words = string.lower().split()

    for word in words:
        result.append(word[0].upper() + word[1:])

    return ' '.join(result)
```

```python
title_case("data science")
```

Q7. Remove all duplicate characters from a string (preserve order).

```python
def remove_dup(string):
  st = ''
  for i in string:
    if i not in st:
      st += i

  return st
```

```python
remove_dup('characters')
```

Q8. Write a function that checks whether two strings are anagrams.

```python
def is_anagrams(string1,string2):
  if len(string1) != len(string2):
    return False
  else:
    freq = {}

    for ch in string1.lower():
        freq[ch] = freq.get(ch, 0) + 1

    for ch in string2.lower():
        if ch not in freq:
            return False

        freq[ch] -= 1
        if freq[ch] < 0:
            return False

    return True
```

```python
is_anagrams('Heart','Earth')
```

Q9. Count the number of words in a sentence.

```python
def count_word(sentence):
    count = 0
    in_word = False

    for ch in sentence:
        if ch != ' ' and not in_word:
            count += 1
            in_word = True
        elif ch == ' ':
            in_word = False

    return count
```

```python
count_word('Ritesh is data scientist')
```

Q10. Write a function that finds the longest word in a sentence.

```python
def longest_word(sen):
    words = sen.split()

    longest = ""
    max_len = 0

    for word in words:
        if len(word) > max_len:
            longest = word
            max_len = len(word)

    return longest
```

```python
longest_word('Ritesh is data scientist')
```
