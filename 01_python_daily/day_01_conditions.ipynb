Q1. Check if number is positive,negative or zero

```python
num = int(input("Enter number: "))
if num > 0:
  print("Number is Positive")
elif num < 0:
  print("Number is Negative")
else:
  print("Number is Zero")
```

Q2. Check if number is Even or Odd

```python
num = int(input("Enter number: "))
if num%2 == 0:
  print(f'{num} is Even')

else:
  print(f'{num} is Odd')
```

Q3. Check if number which number is greater among two integer

```python
num1 = int(input("Enter 1st number: "))
num2 = int(input("Enter 2nd number: "))
if num1 > num2:
  print(f"{num1} is greater")
else:
  print(f"{num2} is greater")
```

Q4. Check if which number is largest among three integers

```python
num1 = int(input("Enter 1st number: "))
num2 = int(input("Enter 2nd number: "))
num3 = int(input("Enter 3rd number: "))

if num1>num2 and num1>num3:
  print(f"{num1} is largest")
elif num2>num1 and num2>num3:
  print(f"{num2} is largest")
else:
  print(f"{num3} is largest")
```

Q5. Check whether number is divisible by both 3 and 5.

```python
num = int(input("Enter number: "))
if num%3 == 0 and num%5 == 0:
  print(f'{num} is divisible by both 3 and 5')

else:
  print(f"{num} is not divisible by both 3 and 5")
```

Q6. Check whether student is Distinction,Pass or Fail

```python
marks = int(input('Enter your marks: '))
if marks>=75:
  print('Distinction')

elif marks>=40 and marks<=74:
  print('Pass')

else:
  print('Fail')
```

Q7. Check whether year is a leap year.

```python
year = int(input('Enter Year: '))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
  print(f'{year} is leap year')

else:
  print(f'{year} is not leap year')
```

Q8. Check whether tempreature Cold,Moderate or Hot

```python
temp = int(input('Enter tempreature: '))
if temp > 30:
  print("Hot")
elif temp >= 15 and temp <= 30:
  print("Moderate")
else:
  print('Cold')
```

Q9. Check whether number is a multiple of 7.

```python
num = int(input('Enter number: '))
if num%7==0:
  print(f'{num} is multiple of 7')

else:
  print(f'{num} is not multiple of 7')
```
