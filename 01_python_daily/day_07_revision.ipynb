Q1. Given a sentence, count word frequency and print the top 2 most frequent words.

```python
sen = input('Enter any sentence: ').lower().split()

word_freq = {}

for i in sen:
  word_freq[i] = word_freq.get(i,0)+1

max_key_1 = max_key_2 = None
max_value_1 = max_value_2 = 0

for key, value in word_freq.items():
    if value > max_value_1:
        max_value_2 = max_value_1
        max_key_2 = max_key_1
        max_value_1 = value
        max_key_1 = key
    elif value > max_value_2 and key != max_key_1:
        max_value_2 = value
        max_key_2 = key

print(f"Top 2 most frequent words: {max_key_1}, {max_key_2}")
```

Q2. Given a list of numbers:

* remove duplicates

* sort it

* return the second largest number

```python
nums = [10, 20, 20, 40, 30, 40]

first = second = float('-inf')

for num in set(nums):
    if num > first:
        second = first
        first = num
    elif num > second:
        second = num

print(second)
```

Q3. Write a function that takes a list and returns:

* sum

* average

* maximum

* minimum

Return as a dictionary.

```python
def aggregate(list1):
  sum = 0
  maximum = float('-inf')
  minimum = float('inf')
  for i in list1:
    sum += i

  for i in list1:
    if i > maximum:
      maximum = i

    if i < minimum:
      minimum = i
  average = sum/len(list1)
  return {'sum':sum,'average':average,'maximum':maximum,'minimum':minimum}
```

```python
aggregate([10, 20, 20, 40, 30, 40])
```

Q4. Given a list of strings, group strings by their first character.

```python
strings = ['apple', 'banana', 'apricot', 'blueberry', 'cherry']

groups = {}

for s in strings:
    first_char = s[0]
    if first_char not in groups:
        groups[first_char] = []
    groups[first_char].append(s)

print(groups)
```

Q5. Check if a string is a palindrome
Ignore case and spaces.

```python
text = input('Enter word: ').lower().replace(' ','')
rev =''
for ch in text:
  rev = ch+rev

if rev == text:
  print("Palindrome")

else:
  print("Not Palindrome")
```

Q8. Given a list of numbers, count how many times each digit (0–9) appears.

```python
nums = [22,13,24,67]

digit_count = {i: 0 for i in range(10)}

for num in nums:
    for digit in str(abs(num)):
        digit_count[int(digit)] += 1

keys_to_remove = [key for key, value in digit_count.items() if value <= 0]
for key in keys_to_remove:
    del digit_count[key]

print(digit_count)
```

Q9. Given student marks dictionary, find students scoring above average.

```python
students = {
    'Ravi': 85,
    'Anita': 92,
    'Suresh': 78,
    'Neha': 95,
    'Kiran': 88
}

total = 0
for marks in students.values():
    total += marks

average = total / len(students)

above_avg = {}

for name, marks in students.items():
    if marks > average:
        above_avg[name] = marks

print(above_avg)
```

Q10. Write a function that:

* takes a sentence

* removes punctuation

* returns the most frequent character

```python
import string

def most_frequent_char(sentence):
    freq = {}

    for ch in sentence.lower():
        if ch not in string.punctuation and ch != ' ':
            freq[ch] = freq.get(ch, 0) + 1

    return max(freq, key=freq.get)
```
