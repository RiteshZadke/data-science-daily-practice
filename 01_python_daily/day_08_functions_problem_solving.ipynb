Q1. Write a function that takes an integer and returns the sum of its digits.

```python
def digit_sum(num):
    num = abs(num)
    total = 0
    while num > 0:
        total += num % 10
        num //= 10
    return total
```

```python
digit_sum(2345)
```

Q2. Write a function that takes a string and returns:

* number of uppercase letters

* number of lowercase letters

Return the result as a tuple.

```python
def count_up_low(text):
  low_count = 0
  up_count = 0
  for i in text.replace(' ',''):
    if i == i.upper() and i not in '1234567890':
      up_count +=1
    if i == i.lower()and i not in '1234567890':
      low_count += 1

  return (low_count,up_count)
```

```python
count_up_low("Data Science Is Cool 23456")
```

Q3. Write a function to check whether a number is prime.

Handle edge cases carefully:

* numbers ≤ 1

* number = 2

Return True or False.

```python
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False

    return True
```

```python
is_prime(73)
```

Q4. Write a function that returns the factorial of a number.

```python
def factorial(num):
    if num < 0:
        return None

    result = 1
    for i in range(1, num + 1):
        result *= i
    return result
```

```python
factorial(6)
```

Q5. Write a function that takes an integer and returns its reverse.

```python
def reverse_number(num):
    sign = -1 if num < 0 else 1
    num = abs(num)

    rev = 0
    while num > 0:
        rev = rev * 10 + (num % 10)
        num //= 10

    return sign * rev
```

```python
reverse_number(234)
```

Q6. Write a function to check whether a string is a palindrome.

```python
def is_palindrome(text):
    text = text.lower().replace(" ", "")

    left = 0
    right = len(text) - 1

    while left < right:
        if text[left] != text[right]:
            return False
        left += 1
        right -= 1

    return True
```

```python
is_palindrome("Never odd or even")
```

Q7. Write a function that returns the second largest number from a list.

Rules:

* Do not use sort() or max()

* Handle duplicate values

```python
def second_largest(lst):
    if len(lst) < 2:
        return None

    largest = None
    second = None

    for num in lst:
        if largest is None or num > largest:
            if largest != num:
                second = largest
            largest = num
        elif num != largest and (second is None or num > second):
            second = num

    return second
```

```python
second_largest([1,2,3,4,5,6,7])
```

Q8. Write a function that counts the number of vowels in a string.

Rules:

* Case-insensitive

* Ignore digits and special characters

```python
def vowel_count(text):
  text = text.lower()
  count = 0
  for ch in text:
    if ch.isalpha():
      if ch in 'aeiou':
        count += 1
  return count
```

```python
vowel_count('Hello World! 123')
```

Q9. Write a function that returns the first n Fibonacci numbers.

```python
def fibonacci_list(n):
    if n <= 0:
        return []
    if n == 1:
        return [0]

    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i-1] + fib[i-2])

    return fib
```

```python
fibonacci_list(6)
```

Q10. Write a function that removes duplicates from a list without changing order.

```python
def remove_dup(list1):
  result = []
  for num in list1:
    if num not in result:
      result.append(num)

  return result
```

```python
remove_dup([1])
```
